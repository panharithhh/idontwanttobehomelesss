```python
import torch
import torch.nn.functional as F 
import matplotlib.pyplot as plt
```

```python
words = open('names.txt', 'r').read().splitlines()
```

```python
words[:8]
```

```python
chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i,s in  enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()}

```

```python
blocksize = 3
X,Y = [], []
# X is the input to nets and Y is the label

for w in words[:5]:

    print(w)
    context = [0] * blocksize 
    
    for ch in w + '.':        
        
        ix = stoi[ch]
        X.append(context)
        Y.append(ix)
        context = context[1:] + [ix]
        
        
X = torch.tensor(X)
Y = torch.tensor(Y)
        
 
```
